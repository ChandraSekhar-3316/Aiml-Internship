```python

import pandas as pd
import numpy as np


nums = [1, 2, 3, 4, 5]
s = pd.Series(nums)

print("Series with Default Index:")
print(s)
```

```python

# ============================
# Creating Pandas Series with Custom Index
# ============================

nums = [1, 2, 3, 4, 5]
s = pd.Series(nums, index=[1, 2, 3, 4, 5])

print("\nSeries with Custom Index:")
print(s)

fruits = ['Orange', 'Banana', 'Mango']
fruit_series = pd.Series(fruits, index=[1, 2, 3])

print("\nFruit Series:")
print(fruit_series)
```

```python

# ============================
# Creating Pandas Series from a Dictionary
# ============================

dct = {
    'name': 'Asabeneh',
    'country': 'Finland',
    'city': 'Helsinki'
}

s = pd.Series(dct)

print("\nSeries from Dictionary:")
print(s)
```

```python
# ============================
# Creating a Constant Pandas Series
# ============================

s = pd.Series(10, index=[1, 2, 3])

print("\nConstant Series:")
print(s)
```

```python

# ============================
# Creating a Pandas Series Using Linspace
# ============================

s = pd.Series(np.linspace(5, 20, 10))

print("\nLinspace Series:")
print(s)
```

```python

# ============================
# Creating DataFrame from List of Lists
# ============================

data = [
    ['Asabeneh', 'Finland', 'Helsinki'],
    ['David', 'UK', 'London'],
    ['John', 'Sweden', 'Stockholm']
]

df = pd.DataFrame(data, columns=['Name', 'Country', 'City'])

print("\nDataFrame from List of Lists:")
print(df)
```

```python

# ============================
# Creating DataFrame Using Dictionary
# ============================

data = {
    'Name': ['Asabeneh', 'David', 'John'],
    'Country': ['Finland', 'UK', 'Sweden'],
    'City': ['Helsinki', 'London', 'Stockholm']
}

df = pd.DataFrame(data)

print("\nDataFrame from Dictionary:")
print(df)
```

```python

# ============================
# Creating a DataFrame for Modification
# ============================

data = [
    {"Name": "Asabeneh", "Country": "Finland", "City": "Helsinki"},
    {"Name": "David", "Country": "UK", "City": "London"},
    {"Name": "John", "Country": "Sweden", "City": "Stockholm"}
]

df = pd.DataFrame(data)

print("\nOriginal DataFrame:")
print(df)
```

```python

# ============================
# Adding New Columns
# ============================

weights = [74, 78, 69]
df['Weight'] = weights

heights = [173, 175, 169]
df['Height'] = heights

print("\nAfter Adding Weight and Height:")
print(df)
```

```python


# ============================
# Calculating BMI
# ============================

def calculate_bmi():
    bmi = []

    for w, h in zip(df['Weight'], df['Height']):
        bmi.append(w / (h * h))

    return bmi


df['BMI'] = calculate_bmi()

print("\nBMI Added:")
print(df)
```

```python
# ============================
# Formatting BMI Column
# ============================

df['BMI'] = round(df['BMI'], 1)

print("\nBMI Rounded:")
print(df)

```

```python
# ============================
# Adding Birth Year and Current Year
# ============================

birth_year = ['1769', '1985', '1990']
current_year = pd.Series(2020, index=[0, 1, 2])

df['Birth Year'] = birth_year
df['Current Year'] = current_year

print("\nBirth Year and Current Year Added:")
print(df)

```

```python

# ============================
# Checking Data Types
# ============================

print("\nWeight Data Type:")
print(df['Weight'].dtype)

print("\nBirth Year Data Type Before:")
print(df['Birth Year'].dtype)

```

```python


# ============================
# Changing Data Types
# ============================

df['Birth Year'] = df['Birth Year'].astype(int)
df['Current Year'] = df['Current Year'].astype(int)

print("\nBirth Year Data Type After:")
print(df['Birth Year'].dtype)

print("\nCurrent Year Data Type:")
print(df['Current Year'].dtype)

```

```python

# ============================
# Calculating Ages
# ============================

df['Ages'] = df['Current Year'] - df['Birth Year']

print("\nDataFrame with Ages:")
print(df)
```

```python

# ============================
# Finding Mean Age (Without Outlier)
# ============================

mean_age = (35 + 30) / 2

print("\nMean Age:", mean_age)
```

```python

# ============================
# Boolean Indexing
# ============================

print("\nPeople Older Than 120:")
print(df[df['Ages'] > 120])

print("\nPeople Younger Than 120:")
print(df[df['Ages'] < 120])
```
