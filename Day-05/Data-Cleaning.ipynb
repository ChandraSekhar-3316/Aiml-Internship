```python
# Step1.generate Messy Data

import pandas as pd
import numpy as np

# Creating a deliberately dirty dataset
data = {
    'Employee_ID': [1, 2, 3, 4, 5, 2, 6, 7],  # 2 is duplicate
    'Salary': [25000, 30000, np.nan, '40000', 35000, 30000, 50000, np.nan],
    'Experience': [2, 4, 3, 5, np.nan, 4, 6, 1],
    'Working_Hours': ['8', '9', 'Unknown', '10', '8', '9', '11', '7']
}

df = pd.DataFrame(data)

print("--- RAW DATA ---")
print(df)
```

```python
# Step 2: Diagnosis
print("\n--- DIAGNOSIS ---")

# 1. Check data types
print("Data Types:\n", df.dtypes)

# 2. Check for missing values
print("\nMissing Values Count:\n", df.isnull().sum())

# 3. Check for exact duplicates
print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
# Step 3:Fix Data type first
print("\n--- CLEANING TYPES ---")

# Convert Salary to numeric
df['Salary'] = pd.to_numeric(df['Salary'])

# Convert Working_Hours to numeric
# 'Unknown' will become NaN
df['Working_Hours'] = pd.to_numeric(
    df['Working_Hours'],
    errors='coerce'
)

print(df.dtypes)
```

```python
# Step 4: Removing Duplicates
print("\n--- HANDLING DUPLICATES ---")

# Keep first occurrence
df = df.drop_duplicates(keep='first')

print(df)
```

```python
# Step 5: Handling Missing Values
print("\n--- HANDLING MISSING VALUES ---")

# Fill missing Salary with median
salary_median = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(salary_median)

# Fill missing Working_Hours with mean
hours_mean = df['Working_Hours'].mean()
df['Working_Hours'] = df['Working_Hours'].fillna(hours_mean)

# Fill missing Experience with median
experience_median = df['Experience'].median()
df['Experience'] = df['Experience'].fillna(experience_median)

print("\n--- FINAL CLEANED DATA ---")
print(df)

print("\nFinal Data Types:\n", df.dtypes)

print("\nFinal Missing Values:\n", df.isnull().sum())
```
