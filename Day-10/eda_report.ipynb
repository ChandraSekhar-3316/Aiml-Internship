1. What are the Top 10 Most Viewed Videos? barh plot



```python
top_views = df.sort_values(
    by='view_count',
    ascending=False,
)[['title', 'view_count']].head(10)

plt.barh(top_views['title'],top_views['view_count'])
plt.title("top views videos")
plt.xlabel('views')
plt.ylabel('title of the video')
plt.show()
```

2. What are the Top 10 Most Liked Videos? barh plot.


```python
top_likes = df.sort_values(
    by='likes',
    ascending=False,
)[['title', 'likes']].head(10)

plt.barh(top_likes['title'],top_likes['likes'])
plt.title("top 10 liked videos")
```

3. Which Videos Received the Most Comments? barh plot.


```python
top_comments = df.sort_values(
    by = "comment_count",
    ascending=False
)[['title','comment_count']].head(10)

plt.barh(top_comments['title'],top_comments['comment_count'])
```

4. Which Videos Received the Most Dislikes? barh plot


```python
most_dislikes = df.sort_values(
    by = "dislikes",
    ascending=False
)[['title','dislikes']].head(10)

plt.barh(most_dislikes['title'],most_dislikes['dislikes'])
```

5. What is the Distribution of Views? histogram plot


```python
plt.hist(df['view_count'], bins=20)

plt.title("Distribution of Views")
plt.xlabel("Views")
plt.ylabel("Frequency")

plt.show()
```

6. Is There a Relationship Between Views and Likes? scatter plot.


```python
import seaborn as sns

sns.scatterplot(
    data = df,
    x='view_count',
    y = 'likes',
)

plt.xlabel('views')
plt.ylabel('likes')
plt.show()

# Doing videos with more views generally receive more likes.
```

10. Which Videos Have the Highest Engagement? barh plot


```python
#create an engagement 
df['engagement'] = (
    df['likes']/df['view_count']*100 + df['comment_count']/df['view_count']*100
)

#create a top dataset
plt.rcParams['font.family'] = ['Nirmala UI','Segoe UI Emoji']
top_engagement = df.sort_values(
    by='engagement',
    ascending=False
)[['title','engagement']].head(10)

plt.barh(top_engagement['title'],top_engagement['engagement'])
plt.xlabel('engagement')
plt.ylabel('title')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("2020-Augustvideos.csv")


# Remove duplicates
df = df.drop_duplicates()

# Check missing values
print(df.isnull().sum())

# Convert numeric columns
numeric_cols = ['view_count', 'likes', 'dislikes', 'comment_count']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)

# Reset index
df.reset_index(drop=True, inplace=True)

# Save cleaned dataset
df.to_csv("cleaned_2020-Augustvideo.csv", index=False)
```

Insights for ML Modeling

Outliers

The dataset contains several outliers in the view_count, likes, dislikes, and comment_count columns. These outliers represent highly viral videos that received exceptionally high engagement compared to other videos. Since these values reflect genuine user behavior and popularity rather than data errors, they were retained. For machine learning models, transformations such as logarithmic scaling can be applied to reduce the impact of extreme values

Collinearity

There is a strong positive relationship between view_count, likes, and comment_count. Videos with higher view counts generally receive more likes and comments. This indicates potential multicollinearity among engagement-related features. Before model training, feature selection or dimensionality reduction techniques may be considered to reduce redundancy.

Feature Importance

The engagement metrics (likes, comments, and views) appear to be the most informative features in the dataset. These variables capture audience interaction and can be useful predictors when estimating video popularity or engagement levels.

Data Distribution


Most engagement-related features are highly right-skewed. A small number of videos achieve extremely high popularity, while the majority receive moderate engagement. Feature scaling and transformation techniques may improve model performance.

Modelling consderations

The dataset is clean and contains no missing values.
No duplicate records were found.
Outliers should be retained because they represent real viral content.
Highly correlated features should be evaluated before model training.

Conclusion 

The dataset contains meaningful engagement signals and is suitable for machine learning tasks such as popularity prediction, engagement forecasting, and trend analysis. Proper handling of skewed distributions and correlated features can help improve model performance and interpretability
